```python
import pandas as pd
import numpy as np
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('accidents 06-20.csv')
df
```

```python
#Bar Chart
fig, ax = plt.subplots(1,figsize=(24,6))
plt.barh(df.Year, df.Accidents,)
plt.title('Fatalities from 2006 - 2020')
plt.xlabel('Year', fontsize=14)
plt.ylabel('Accidents', fontsize=14)
```

```python
sns.lineplot(data = df, x='Year', y='Accidents')
sns.set_style("whitegrid")
plt.title("Fatalities from 2006 to 2020", fontsize=12)
```

```python
line_chrt = sns.relplot(data = df, x='Year', y ='Accidents', hue = "Accidents")
line_chrt = plt.title("Fatalities from 2006 to 2020")
line_chrt
```

```python
#Step Chart
plt.step(data = df, x='Year', y ='Accidents')
plt.title("Fatalities from 2006 to 2020")
```
